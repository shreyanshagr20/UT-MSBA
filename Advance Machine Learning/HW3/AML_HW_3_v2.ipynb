# Question 2 : t-SNE (7 pts)

Read this [article](https://erdem.pl/2020/04/t-sne-clearly-explained) on t-SNE, a dimensionality reduction technique for visualization and explain it in your own words in one or two paragraphs.

# Answer:

t-SNE is a non linear dimensionality reduction method which is used to understand high-dimensional data which is linearly inseparable.

How it works?
* Algorithm firstly creates a probability distribution which represents the similarity between the data points
* Similarity is defined as the conditional probability that a data point "I" would pick "j" as its neighbor
* First Euclidian distance b/w the data point and its neighbors are calculated. This distance is proportional to the probability density under a Gaussian centered at "i"
* Few more steps are performed to account for high cluster distance, and to make the sum of probability to 1  
* So, now how do we determine the variance of the gaussian distribution? It is determined by the term called "perplexity". It is loosely defined as the number of target neighbors around the central point. SNE uses binary search to get the value of the sigma that generates a probability distribution using the perplexity specified by the user, typically it ranges between 5-50.

Creating low dimensional space
* Now the idea of t-SNE is to translate the original gaussian distribution in a lower dimensional space using the same number of data point.
* We are not using gaussian distribution here because it has a short tails which leads to crowding problem. Instead, a student t-distribution with a single degree of freedom is used
* It provides an advantage that it falls quickly and has a long tail which prevents crowing

Optimizing the distribution of t-SNE using Gradient Descent
* To optimize this distribution t-SNE is using Kullback-Leibler divergence between the conditional probabilities p i|j and q i|j‚Äã
* A gradient is calculated for each point and describes how ‚Äústrong‚Äù it should be pulled and the direction it should choose
	
Optimizing the process
* Early Compression: To avoid clustering at a early stage, t-SNE is adding a L2 penalty to the cost function to avoid algorithm not to focus on the local groups. 
* Early Exaggeration: We multiply Pi|j in early stages which enables moving clusters of Qi|j more, as a result clusters don't get in each other's way

Use in CNN
* Deep CNN networks are black box models, and its becomes impossible to interpret the layers given they are in a very high dimension. 
* But using t-SNE, we can create maps to display which input data seems "similar" for the network

Although it‚Äôs an iterative, non-deterministic process, and can't be used for ML model training purposes, it is a super powerful technique to understand high-dimensional datasets and is used widely in the industry#### Answer:

# Question 3: PCA (8 pts)

Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



# Answer

PCA is a feature extraction method that is aimed to reduced the dimensionality of the feature space. Broadly speaking, PCA projects the data from a n dimensional space to a m dimensional space (where m<n) by linearly combining the n features so that maximum amount of variance is retained. Below is the algorithm of the same - 

* A typical way to do PCA is to break the original n dimensional space into eigenvalues and eigenvectors through Singular Value Decomposition (SVD)
* Suppose your original vector is X, then X . X(transpose)=P . D . P(inverse) such that P is your eigenvector and D contains eigenvalues at the diagonal elements and rest all elements in D are zero. And X . X(transpose) is your covariance matrix. The eigenvalues on the diagonal of D will be associated with the corresponding column in P i.e., the first element of D is Œª‚ÇÅ and the corresponding eigenvector is the first column of P
* Now you need to sort the eigen values from largest to smallest and correspondingly sort your eigen vectors. You now choose the top m values.
* Now, X* = X.P* where P* is the matrix that contains the top m eigen vectors as columns. X* is your representation of data (or projection) in the m dimensional space (i.e. your transformed data)

Now the eigenvectors represent the directions and the eigenvalues loosely speaking indicate how much variance is explained in the direction of the eigen vector. We select the top eigenvalues implying that we are selecting the top eigenvectors (or directions) which explain the maximum variance in the data. Which means the first principal component will explain the maximum variance, the 2nd will explain the end highest variance and so on. Idea of PCA is to represent a "n" dimensional space in "m" dimensions without much loss of information (i.e. variance), so selecting the eigenvectors the above described way is the optimal choice as it ensures that maximum variance is retained for a given value of m and the reconstruction loss is the lowest.

The way you reconstruct (a noisy version of) the original data is by taking the dot product of the transpose matrix containing the subset("m") of eigen vectors with the transformed data. i.e. X (noisy version) = P*(transpose) . X*

e) i)**(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

ii)**(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
import scipy.stats as st

np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value
sample_std = sample.std()
z_critical = round(st.norm.ppf(.975), 2)

#Get population standard deviation
pop_stdev = data_dm["SalePrice"].std()


#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print("True mean value is {mean}\n95% confidence interval is: {lower_bound} - {upper_bound}".format(mean = sample_mean,
                                                                                            lower_bound = round(confidence_interval[0],5),
                                                                                            upper_bound = round(confidence_interval[1],5)))

````

````python
# Part (ii)
np.random.seed(12)

sample_size = 500

intervals = []
sample_means = []

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    # compute z critical value
    sample_std = sample.std()
    z_critical = round(st.norm.ppf(.975), 2) #95%                       

    #Get population standard deviation
    pop_stdev = data_dm["SalePrice"].std()

    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    
    intervals.append(confidence_interval)
    
plt.figure(figsize=(13, 9))

plt.errorbar(x=np.arange(0.1, 100, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for bot,top in intervals],
             fmt='o')

# plt.hlines(xmin=0, xmax=25,
plt.hlines(xmin=0, xmax=100,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````


As per the above chart, the red line indicates the population mean and the vertical lines depict the sample mean and their respective 95% Confidence intervals(i.e. 1.96*Stdev) for the 100 trials

We observe that, out of the 100 samples from the population only 3 samples lie outisde the aloowed deviation.
