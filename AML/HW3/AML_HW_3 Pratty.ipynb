# Question 2 : t-SNE (7 pts)

Read this [article](https://erdem.pl/2020/04/t-sne-clearly-explained) on t-SNE, a dimensionality reduction technique for visualization and explain it in your own words in one or two paragraphs.

#### Answer:

#### Answer:

t-SNE is an unsupervised non-linear dimensionality reduction technique which is primarily used for data visualization to project high-dimensional data onto lower dimensional space, instead of ML training since it cannot be reapplied in the same way. It is stocahstic in nature and works in an iterative fashion so it gives different results each time you run it. It finds appplication in visualizing high-dimensional data like CNN feature maps which are particularly tricky to interpret.

To summarize the steps involved, it calculates scaled similarities for each point through a normal probability distribution such that neighbors in high dimension have higher similarity. Then, all points are initially plotted in the lower dimension randomly, calculate scaled similarity using t-distribution (to avoid clumping of cluster basically), then optimize these new similarities in lower dimension to be closer to high dimension similarities. Gradient Descent is performed by moving the points in low dimension space, calculating gradient and determining where to move the points as per this gradient. Essentially, for a given point, points from similar cluster attract each other and those from different clusters repel it. The result is a projection of original data points in a lower dimension (with majority of original information captured) that can be visualized.

#Question 3: PCA (8 pts)

Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



#### Answer:

Principal Component Analysis (PCA) is an unsupervised linear transformation technique that is widely used across many fields. It helps to find the directions of maximum variance in high dimensional data and projects it onto a new subspace with equal or fewer dimensions than the original one.

The Eigenvectors are essentially axes in the new subspace try to capture maximum variation in the data. Each eigenvector is a linear combination of original features, and we can choose what number of eigenvectors to use to represent the data based on how much variation each eigenvector captures (by looking at corresponding eigenvalues and explained variance). So it is optimal to use eigenvectors to represent the data since they are simply projections of original data onto lower dimension.

Original data can be reconstructed from the eigenvectors and loadings (PC scores) by the below matrix formula -

PCA reconstruction = PC scores * Transpose(Eigenvectors) + Mean

where Mean is the mean vector we used to center the data during PCA.

If we use less number of PCs, the reconstruction will be noisy.

````python
import pandas as pd
# Load data code goes here
train_df = pd.read_csv('Q3_train_dataset.csv')
test_df = pd.read_csv('Q3_test_dataset.csv')

X_train = train_df[['X_train_0', 'X_train_1', 'X_train_2', 'X_train_3', 'X_train_4']].to_numpy()
y_train = train_df[['y_train']].to_numpy()
X_test = test_df[['X_test_0', 'X_test_1', 'X_test_2', 'X_test_3', 'X_test_4']].to_numpy()
y_test = test_df[['y_test']].to_numpy()
````

````python
# (a)
from pyod.models.knn import KNN
clf = KNN() # initialize KNN class using the default parameters

# fit the model using X_train

knn_fit = clf.fit(X_train)
````

````python
from pyod.utils.data import evaluate_print
# predict the outlier labels of X_test using the trained model, compute the raw outlier scores on X_test using decision_function()
# then use evaluate_print() to print out the evaluation results

evaluate_print(clf_name = 'KNN', y=y_test, y_pred=knn_fit.decision_function(X_test))
````

````python
# (b)
# this has some issue with the dimensions - check with TA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

# Fit pca to X_train and X_test and transform 
train_principalComponents = pca.fit_transform(X_train)
test_principalComponents = pca.fit_transform(X_test)

knn_pca_fit_train = clf.fit(train_principalComponents)
knn_pca_fit_test = clf.fit(test_principalComponents)

from pyod.utils.example import visualize
# Visualize the ground truth outliers and predicted outliers using visualize()

visualize(clf_name='KNN', X_train=train_principalComponents, y_train=y_train, 
          X_test=test_principalComponents, y_test=y_test, y_train_pred=knn_pca_fit_train.predict(train_principalComponents), 
          y_test_pred=knn_pca_fit_test.predict(test_principalComponents))
````

````python
# (c)
from pyod.models.inne import INNE

# Fit pca to X_train and X_test and transform 
train_principalComponents = pca.fit_transform(X_train)
test_principalComponents = pca.fit_transform(X_test)

inne_clf = INNE()
inne_pca_fit_train = inne_clf.fit(train_principalComponents)
inne_pca_fit_test = inne_clf.fit(test_principalComponents)

from pyod.utils.example import visualize
# Visualize the ground truth outliers and predicted outliers using visualize()

visualize(clf_name='INNE', X_train=train_principalComponents, y_train=y_train, 
          X_test=test_principalComponents, y_test=y_test, y_train_pred=inne_pca_fit_train.predict(train_principalComponents), 
          y_test_pred=inne_pca_fit_test.predict(test_principalComponents))
````

````python
print("Performance of KNN")
evaluate_print(clf_name = 'KNN', y=y_test, y_pred=knn_pca_fit_test.decision_function(test_principalComponents))

print("Performance of INNE")
evaluate_print(clf_name = 'INNE', y=y_test, y_pred=inne_pca_fit_test.decision_function(test_principalComponents))

# KNN seems to be performing better than INNE in our case
````

e) i)**(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

ii)**(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()

#Get the critical Z value
# 95% confidence - 100 - 95 = 5%
# 0.05/2 = 0.025
# corresponding z-value for 0.025 =  1.65 (from the table)
z_critical = 1.65

#Get population standard deviation
pop_stdev = data_dm['SalePrice'].std()


#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

#Print confidence interval and true mean value
print(confidence_interval, data_dm['SalePrice'].mean())

````

````python
np.random.seed(12)

#define the sample size
sample_size = 500
intervals = []
sample_means = []

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    #compute z critical value
    z_critical = 1.65

    #compute population std dev        
    pop_stdev = data_dm['SalePrice'].std()
    
    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error)  
    
    intervals.append(confidence_interval)
    

plt.figure(figsize=(13, 9))

plt.errorbar(x=np.arange(0.1, 100, 1), 
             y=sample_means, 
             yerr=[(top-bot)/2 for top,bot in intervals],
             fmt='o')

plt.hlines(xmin=0, xmax=25,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````

### Solution:
The mean of the sampling distributions is approximately equal to the population mean. This is true because of the central limit theorem.
